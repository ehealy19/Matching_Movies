```python
import pandas as pd
import numpy as np

movies_df = pd.read_csv("../data/output_data/clean_movies_sent_fem_star.csv")

print(movies_df.head(5))
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm

movies_df.columns = movies_df.columns.str.replace(' ', '_').str.replace('[^0-9a-zA-Z_]', '', regex=True)
response_var = 'Female_Lead'
predictor_vars = [col for col in movies_df.columns if col != response_var]
formula = f"{response_var} ~ " + " + ".join(predictor_vars)

prop_model = smf.glm(
    formula=formula,
    data=movies_df,
    family=sm.families.Binomial()
).fit()

print(prop_model.summary())
```

```python
movies_df['ps'] = prop_model.predict(movies_df)
movies_df
```
